```python
from tqdm import tqdm
from skimage.io import imread
from skimage.io import imshow
from skimage.transform import resize
from tensorflow import keras
from keras.utils import array_to_img, img_to_array
import numpy as np
import matplotlib.pyplot as plt
import os
IMG_SIZE=128
IMG_SHAPE=128
NUM_CHANNELS=3

# Assuming these variables are defined correctly


TRAIN_PATH="/content/drive/MyDrive/data-science-bowl-2018/stage1_train/"
TEST_PATH="/content/drive/MyDrive/data-science-bowl-2018/stage1_test/"


train_id=next(os.walk(TRAIN_PATH))[1]
test_id=next(os.walk(TEST_PATH))[1]


X_train = np.zeros((len(train_id), IMG_SIZE, IMG_SHAPE, NUM_CHANNELS), dtype=np.uint8)
y_train = np.zeros((len(train_id), IMG_SIZE, IMG_SHAPE, 1),dtype=np.bool_)



for i,id_ in tqdm(enumerate(train_id),total=len(train_id)):
  path= TRAIN_PATH +id_
  img=imread(path+"/images/"+id_+".png")[:,:,:NUM_CHANNELS] #take file name)
  img=resize(img,(IMG_SIZE,IMG_SHAPE),mode='constant',preserve_range=True)
  #print((img).shape)
  X_train[i]=img
  #print(X_train[i].shape)
  #imshow(X_train[i])
  mask_=np.zeros((IMG_SIZE,IMG_SHAPE,1),dtype=np.bool_)
  for mask_file in (next(os.walk(path+"/masks"))[2]):
      mask=imread(path+"/masks/"+mask_file)
      mask=resize(mask,(IMG_SIZE,IMG_SHAPE),mode='constant',preserve_range=True)
      mask=np.expand_dims(mask,axis=-1)

      mask_=np.maximum(mask,mask_)
  y_train[i]=mask_
  imshow(y_train[i])
  plt.axis('off')
  plt.show()

#resizing testing images


```

```python
X_train,y_train
import tensorflow
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Conv2DTranspose, Dropout, Input, concatenate, UpSampling2D
from keras.utils import plot_model
import tensorflow as tf
import matplotlib.pyplot as plt
import random

input_shape=(128,128,3)

inputs = Input(input_shape)
inputs=tf.keras.layers.Rescaling(1./255)(inputs)

# Downsampling path
c1 = Conv2D(16, (3, 3), activation='relu', padding='same')(inputs)
c1 = Dropout(0.1)(c1)
c1 = Conv2D(16, (3, 3), activation='relu', padding='same')(c1)
p1 = MaxPooling2D((2, 2))(c1)
print(p1.shape)
print()

c2 = Conv2D(32, (3, 3), activation='relu', padding='same')(p1) #Includes some weight initalizqation steps as well (Refer CampusS)
c2 = Dropout(0.1)(c2)
c2 = Conv2D(32, (3, 3), activation='relu', padding='same')(c2)
p2 = MaxPooling2D((2, 2))(c2)
print(p2.shape)
print()


c3 = Conv2D(64, (3, 3), activation='relu', padding='same')(p2)
c3 = Dropout(0.1)(c3)
c3 = Conv2D(64, (3, 3), activation='relu', padding='same')(c3)
p3 = MaxPooling2D((2, 2))(c3)
print("This is c3")
print(c3.shape)
print("This is p3 (After maxpooling- reduces the shape by half)")
print(p3.shape)
print()


c4 = Conv2D(128, (3, 3), activation='relu', padding='same')(p3)
c4 = Dropout(0.1)(c4)
c4 = Conv2D(128, (3, 3), activation='relu', padding='same')(c4)
p4 = MaxPooling2D((2, 2))(c4)
print(p4.shape)
print()


c5 = Conv2D(256, (3, 3), activation='relu', padding='same')(p4)
c5 = Dropout(0.1)(c5)
c5 = Conv2D(256, (3, 3), activation='relu', padding='same')(c5)
print(c5.shape)
print()
print("Upsampling start: ")


# Upsampling path

c6=Conv2DTranspose(128,(2,2),strides=(2,2),activation='relu',padding='same')(c5)
print(c6.shape)
print(c4.shape)
c6=concatenate([c6,c4])
print(c6.shape)
#downsizing along with input features

c7=Conv2D(128,(3,3),activation='relu',padding='same')(c6)
c7=Dropout(0.1)(c7)
c7=Conv2D(128,(3,3),activation='relu',padding='same')(c7)
print()
print(c7.shape)


c6=Conv2DTranspose(128,(2,2),strides=(2,2),activation='relu',padding='same')(c5)
print(c6.shape)
print(c4.shape)
c6=concatenate([c6,c4])
print(c6.shape)
print("New from here")
#downsizing along with input features

c7=Conv2D(128,(3,3),activation='relu',padding='same')(c6)
c7=Dropout(0.1)(c7)
c7=Conv2D(128,(3,3),activation='relu',padding='same')(c7)
print(c7.shape)


c8=Conv2DTranspose(64,(2,2),strides=(2,2),activation='relu',padding='same')(c7)
print(c8.shape)
print(c3.shape)
c9=concatenate([c8,c3])
print(c9.shape)
#downsizing along with input features
c10=Conv2D(64,(3,3),activation='relu',padding='same')(c9)
c11=Dropout(0.1)(c10)
c12=Conv2D(64,(3,3),activation='relu',padding='same')(c11)
print(c12.shape)

c13=Conv2DTranspose(32,(2,2),strides=(2,2),activation='relu',padding='same')(c12)
c14=concatenate([c13,c2])
c15=Conv2D(32,(3,3),activation='relu',padding='same')(c14)
c16=Dropout(0.1)(c15)
c17=Conv2D(32,(3,3),activation='relu',padding='same')(c16)
print(c17.shape)

c17=Conv2DTranspose(16,(2,2),strides=(2,2),activation='relu',padding='same')(c17)
c18=concatenate([c17,c1])
c19=Conv2D(16,(3,3),activation='relu',padding='same')(c18)
c20=Dropout(0.1)(c19)
c21=Conv2D(16,(3,3),activation='relu',padding='same')(c20)
print(c21.shape)
output=Conv2D(1,(1,1),activation='sigmoid')(c21) #Wht one output for binary segmentation?---
print("This is the shape of the output")
print(output.shape)
model=keras.Model(inputs=[inputs],outputs=[output])
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
#plot model
#For an output with more channels - maybe change the loss fucntion from binary crossentrophy to sparse cross entrophy loss
#Why binary cross entrophy is used as a loss function?
checkpointer=keras.callbacks.ModelCheckpoint('unet.h5',save_best_only=True)

callbacks=[
  keras.callbacks.EarlyStopping(patience=3,restore_best_weights=True,monitor="val_loss"),
  keras.callbacks.TensorBoard(log_dir='./logs'),
  checkpointer
]
results=model.fit(X_train,y_train,epochs=100,batch_size=4,callbacks=callbacks,validation_split=0.1)
model.summary()
```
